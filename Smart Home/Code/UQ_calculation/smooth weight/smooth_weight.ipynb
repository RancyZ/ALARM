```python
import pandas as pd
import numpy as np
from scipy.ndimage import gaussian_filter1d

# New data
df = pd.DataFrame({
    "P": [5, 10, 15, 20, 25, 30, 35, 40],

    #manually change the original optimal weight to your actual optimal weights
    "alpha1": [0, 0.2, 0.2, 0.2, 0.2, 0, 0, 0],
    "alpha2": [0.8, 0.8, 0.8, 0.4, 0.4, 1, 0.9, 0.8],
    "alpha3": [0.2, 0, 0, 0.4, 0.4, 0, 0.1, 0.2]
})

# Gaussian smoothing
sigma = 1.3
df_smooth = df.copy()
for col in ["alpha1", "alpha2", "alpha3"]:
    df_smooth[col] = gaussian_filter1d(df[col], sigma=sigma)

# Renormalize so each row sums to 1
df_smooth[["alpha1", "alpha2", "alpha3"]] = (
    df_smooth[["alpha1", "alpha2", "alpha3"]]
    .div(df_smooth[["alpha1", "alpha2", "alpha3"]].sum(axis=1), axis=0)
)

# Save to Excel
df_smooth.to_excel(
    "best_alpha_combination_for_each_P_meansigma_train_all_folds_smooth.xlsx",
    index=False
)
```
